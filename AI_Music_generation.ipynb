```python
#Preprocess all files
```

```python
notes = []
for file in glob.glob("./midi_songs/*.mid"):
    midi = converter.parse(file) #converts to stream.score object
    print("Parsing %s"%file)
    elements_to_parse = midi.flat.notes
    for ele in elements_to_parse:
        #If the element is a note, them store its pitch
        if isinstance(ele, note.Note):
            notes.append(str(ele.pitch))
        #If element is a chord, split them and them with +
        elif isinstance(ele, chord.Chord):
            notes.append("+".join(str(n) for n in ele.normalOrder))
    
```

```python
len(notes)
```

```python
with open("notes","wb") as f:
    pickle.dump(notes, f)
```

```python
with open("notes", "rb") as f:
    notes = pickle.load(f)
```

```python
len(notes)
```

```python
len(set(notes))
```

```python
n_vocab = len(set(notes))
```

```python
#Number of elements one LSTM layer with consider
sequence_length = 100
```

```python
pitch_names = sorted(set(notes))
```

```python
#Mapping between elements to int values
ele_to_int = {}
for idx, ele in enumerate(pitch_names):
    ele_to_int[ele] = idx
```

```python
ele_to_int
```

```python
network_input = []
network_output = []

for i in range(len(notes) - sequence_length):
    seq_in = notes[i:i+sequence_length]
    seq_out = notes[i+sequence_length]
    network_input.append([ele_to_int[ch] for ch in seq_in])
    network_output.append(ele_to_int[seq_out])
```

```python
network_output[0]
```

```python
#no of examples
n_pattern = len(network_input)
print(n_pattern)
```

```python
from keras.utils import to_categorical
```

```python
network_output = np.array(network_output)
```

```python
print(network_output.shape)
net_output = to_categorical(network_output)
print(net_output.shape)
```

```python
network_input = np.array(network_input)
network_input = network_input.reshape(network_input.shape[0], network_input.shape[1], 1)
```

```python
norm_network_input = network_input/float(n_vocab)
```

```python
print(norm_network_input.shape)
```

```python
from keras.models import Sequential, Model
from keras.layers import *
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
model=Sequential()
model.add(LSTM(512, input_shape = (norm_network_input.shape[1], norm_network_input.shape[2]), return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(512,  return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(512,  return_sequences = False))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(n_vocab, activation = "softmax"))
model.compile(loss = "categorical_crossentropy", optimizer = "adam")
model.summary()
```

```python
checkpoint = ModelCheckpoint("best_model.hdf5" ,monitor = "loss", mode = "min", save_best_only=True)
history = model.fit(norm_network_input, net_output, epochs = 100, batch_size = 64, callbacks = [checkpoint])
```

```python
from keras.models import load_model
saved_model = load_model('best_model.hdf5')
```

```python
test_input = network_input.reshape(-1,100)
```

```python
start_idx = np.random.randint(10000)
print(start_idx)
```

```python
test_input[start_idx]
```

```python
int_to_ele = {}
for ele, int_ in ele_to_int.items():
    int_to_ele[int_] = ele
```

```python
int_to_ele
```

```python
ele_to_int
```

```python
pat = test_input[start_idx]
pattern = [ele for ele in pat]
print
pred_output = []
#generate 200 elements
for note_idx in range(200):
    pred_input = np.reshape(pattern, (1,len(pattern),1))
    pred_input = pred_input/float(n_vocab)
    pred = saved_model.predict(pred_input, verbose = 0)
#     print(pred_out)
    idx = np.argmax(pred)
#     print(int_to_ele[idx])
    result = int_to_ele[idx]
    pred_output.append(result)
    pattern.append(idx)
    pattern = pattern[1:]
```

```python
pred_output
```

```python
#Create midi file
```

```python
offset = 0 #Time
output_notes = []
for p in pred_output:
    if ("+" in p) or p.isdigit():
        notes_in_chord = p.split("+")
        temp_notes = []
        for cur_note in notes_in_chord:
            new_note = note.Note(int(cur_note))
            new_note.instrument = instrument.Piano()
            temp_notes.append(new_note)
        new_chord = chord.Chord(temp_notes)
        new_chord.offset = offset
        output_notes.append(new_chord)
    else:
        new_note = note.Note(p)
        #print(new_note)
        new_note.offset = offset
        new_note.stored_instrument = instrument.Piano()
        output_notes.append(new_note)
    
    offset += 0.5
            
```
